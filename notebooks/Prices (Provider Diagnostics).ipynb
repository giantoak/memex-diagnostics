```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pylab
import seaborn
import os

root_dir = '/mnt/memex-diagnostics'
data_dir = os.path.join(root_dir, 'data/dump4')
plot_dir = os.path.join(root_dir, 'plots/intraprovider-prices')

prices_by_ad = pd.read_csv(os.path.join(data_dir, 'made/ad_prices_by_ad.csv'))
f = open(os.path.join(data_dir, 'made/doc-provider-sorted.tsv'))
```

```python
f.close()
```

```python
# Create a mapping from ad_ids to clusters
mappings = {}
for line in f:
    parts = line.strip().split(',')
    mappings[parts[2]] = parts[0]
```

```python
# Get clusters
prices_by_ad['cluster'] = prices_by_ad.apply(lambda x: mappings.get(x.ad_id, ''), 1)
prices_by_ad.shape, prices_by_ad.head()
```

```python
# Drop ads without any cluster or price
print (prices_by_ad.cluster == '').sum()

prices_by_ad = prices_by_ad[prices_by_ad.cluster != '']
prices_by_ad.dropna(subset=['price_per_hour'], inplace=True)
print prices_by_ad.shape
```

```python
# Filter clusters based off some attributes:
# - must have > 5 members
# - must be a valid cluster identifier (they all start with "cluster:")

MIN_CLUSTER_SIZE = 5

clusters = prices_by_ad.groupby('cluster')
df = clusters.filter(lambda x: len(x) > MIN_CLUSTER_SIZE)
```

```python
print df.head()
```

```python
df.groupby('cluster').count().ad_id.sort(inplace=False)
```

```python
clusters = df.groupby('cluster')

print "Number of clusters: {}. Number of ads: {}".format( clusters.ngroups, df.shape[0] )
```

```python
def draw_kde(column, x):
    ''' Helper function for plotting kde charts '''
    pylab.clf()
    column.plot(kind='kde')
    axes = pylab.gca()
    axes.set_xlim(x)
    
draw_kde(clusters.price_per_hour.std(), [0, 90])
```

```python
clusters.price_per_hour.std().describe()
```

```python
# Set up scatterplot between standard deviation and cluster size
both = clusters.agg({'price_per_hour': [np.std, len]})['price_per_hour']

pylab.scatter(both['len'], both['std'], alpha=0.3)
pylab.xlabel('# of ads in provider')
pylab.ylabel('std price per hour')
pylab.title('Standard deviations of intraprovider prices by provider size')
pylab.savefig(os.path.join(plot_dir, 'std_prices_by_cluster_size.png'))
```

```python
# With log scale
pylab.scatter(np.log(both['len']), np.log(both['std']+1), alpha=0.3)
pylab.xlabel('Log # of ads in provider')
pylab.ylabel('Log std price per hour')
pylab.title('Log standard deviations of intraprovider prices by log provider size')
pylab.savefig(os.path.join(plot_dir, 'log_std_prices_by_cluster_size.png'))
```

```python
# With outlier removal
std_size = both['len'].std()
std_std = both['std'].std()
print "std(std): {}, std(len): {}".format(std_std, std_size)

both_pruned = both[(both['len'] < std_size*3) & (both['std'] < std_std*3)]

print "Data points before pruning: {}, after pruning: {}".format(len(both), len(both_pruned))

pylab.scatter(both_pruned['len'], both_pruned['std'], alpha=0.3)
pylab.xlabel('# of ads in provider')
pylab.ylabel('std price per hour')
pylab.title('Standard deviations of intraprovider prices by log provider size (without outliers)')
pylab.savefig(os.path.join(plot_dir, 'std_prices_by_cluster_size_no_outliers.png'))
```

```python
#With log scale and outliers removed

pylab.scatter(np.log(both_pruned['len']), np.log(both_pruned['std']+1), alpha=0.3)
pylab.xlabel('Log # of ads in provider')
pylab.ylabel('Log std price per hour')
pylab.title('Log standard deviations of intraprovider prices by log provider size (without outliers)')
pylab.savefig(os.path.join(plot_dir, 'log_std_prices_by_cluster_size_no_outliers.png'))
```
